```python
#Recommend user 3 a job based on another user who has almost the same skill as him.

print("\nRecommending jobs to USER 3 using collaborative filtering\n")
m1=max(sim[3][:3])
m2=max(sim[3][4:])
ma=max(m1,m2)
suser=sim[3].index(ma)
print("User ",suser," was most similiar to User 3 on comparing skills\n")
print("Using collaborative filtering, user ",suser,"'s jobs can be recommendeded to user 7 ")
print("The recommended companies are:\n")
print(dfcont.loc[dfcont.Respondent==suser]["company"].values)
print("The recommended job titles are:\n ")
print(dfcont.loc[dfcont.Respondent==suser]["jobtitle"].values)
```
